```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dimd=sns.load_dataset('diamonds')
```

```python
dimd.head()
```

```python
dimd=dimd.sample(n=200,random_state=44)
```

```python
dimd.head()
```

# basic

```python
sns.set_style('white')
```

```python
sns.regplot(x='carat'	,y='price',data=dimd)
```

```python
#remove the line 
sns.regplot(x='carat'	,y='price',data=dimd,
            fit_reg=False)
```

```python
# remove scatter
sns.regplot(x='carat'	,y='price',data=dimd,
            scatter=False)# the shaded portion mear the line are confidence intervals it is created by using bootstrapping
```

```python
# removethe confidence interval
sns.regplot(x='carat'	,y='price',data=dimd,ci=None)
```

## Descrete variables

```python
# mapping the cut value of diamond with numerical values
cut_map = {
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Premium': 4,
    'Ideal': 5
}

dimd['cut_value'] = dimd.cut.map(cut_map)
```

```python
dimd.cut_value.value_counts()
```

```python
dimd.dropna(inplace=True)
```

```python
dimd.head()
```

```python
#jitter
sns.regplot(x='depth',y='price',data=dimd,x_jitter=0.1)
```

```python
#estimators
import numpy as np

sns.regplot(x='depth',y='price',data=dimd,
            x_estimator=np.mean)
```

## robust regression

```python
x_example=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
y_example=[0.1, 0.8, 2.2, 2.7, 3.8, 4.5, 6.2, 6.8, 7.9, 9.4, 30.4]
```

```python
sns.regplot(x=x_example, 
            y=y_example,
            ci=None
            )# there is an outlier in the plot
```

```python
# robust : to make the regeression model robust to outliers
sns.regplot(x=x_example, 
            y=y_example,
            ci=None,
            robust=True
            )
```

## other regressions
1. order(polynomial regression)

2. robust(robust form of linear regression helps to control outliers)

3. logistic(use whe one of the variable become binary)

4. lowess(locally weighted scatter plot smoothing:form of regerssiom allos you to produce smoothed out curve)

they are mutually exclusive ie, can only chose one at a time


```python
sns.regplot(x=x_example, 
            y=y_example,
            ci=None,
            lowess=True
            )
```
