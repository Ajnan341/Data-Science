```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

it is a plot to count the number of observations you have per category and display tha information as bars

```python
dimd=sns.load_dataset('diamonds').dropna()
```

```python
dimd.head()
```

```python
dimd=dimd[dimd.clarity.isin(['SI1','VS2'])]
```

```python
dimd.shape
```

```python
dimd.head()
```

```python
sns.set_style('dark')
```

## Basics

```python
sns.countplot(x='color',data=dimd)
```

```python
sns.countplot(x='color',data=dimd)
```

```python
dimd.cut.cat.categories
```

```python
#change the order 
odr=['J','D','E','I','F','H','G']
sns.countplot(x='color',data=dimd)
```

```python
#sorting valuses
sns.countplot(x='color',data=dimd,order=dimd.color.value_counts().index)
```

```python
sns.countplot(x='color',data=dimd,order=dimd.color.value_counts().index[::-1])
```

Hue

```python
sns.countplot(x='color',data=dimd,hue='clarity')
```

```python
sns.countplot(x='color',data=dimd,hue='clarity',hue_order=['SI1','VS2'])
```

## Styling

```python
sns.countplot(x='color',data=dimd,hue='clarity',palette='hot')
```

```python
sns.countplot(x='color',data=dimd,lw=10)
```

```python
sns.countplot(x='color',data=dimd,lw=10,edgecolor='r')
```
