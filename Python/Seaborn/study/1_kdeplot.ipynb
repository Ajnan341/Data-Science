```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars=sns.load_dataset('mpg')
cars
```

```python
cars.dropna(inplace=True)
```

```python
cars.info()
```

```python
cars.describe()
```

## kernel density estimation plot

```python
sns.kdeplot(cars.horsepower)
```

```python
#shadding
sns.kdeplot(cars.horsepower,fill=True)
```

```python
#bandwidth
sns.kdeplot(cars.horsepower,bw_adjust=10)
```

The bw_adjust parameter is a scalar multiplier that adjusts the bandwidth used by the KDE. It scales the bandwidth calculated by the default method or the method specified by bw_method.

```python
sns.kdeplot(cars.horsepower,bw_adjust=0.2)
```

```python
sns.kdeplot(cars.horsepower,bw_method=10)
```

The bw_method parameter specifies the method used to calculate the bandwidth of the kernel. This parameter can take a string indicating a predefined method or a function that computes the bandwidth.

```python
sns.kdeplot(cars.horsepower,bw_method=0.2)
```

CUMULATIVE

Using cumulative=True effectively allows you to understand how data accumulates up to certain points, making it a powerful tool for exploratory data analysis.

```python
sns.kdeplot(cars.horsepower,cumulative=True)
```

```python
sns.kdeplot(cars.horsepower,cumulative=False)
```

## Bivariate


```python

sns.kdeplot(x=cars.horsepower,y=cars.mpg)
```

```python
#adding more levels
sns.kdeplot(x=cars.horsepower,y=cars.mpg,n_levels=20)
```

```python
#shadding
sns.kdeplot(x=cars.horsepower,y=cars.mpg,fill=True)
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg,fill=True,thresh=False)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x=cars.horsepower,y=cars.mpg,fill=True,cbar=True)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x=cars.horsepower,y=cars.mpg,fill=True,cbar=True,color='r')
```

```python
#combination of two attribute (univariate)
sns.kdeplot([cars.horsepower,cars.mpg],fill=True,cbar=True,color='r')
```
