```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

Lmplot stands for linear model plot with it we can create linear models for data like regplot and also has facetgrit to create multiple regplots

```python
df = sns.load_dataset('penguins').dropna()

df.head()
```

```python
sns.set_style('white')
```

# Basic


```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df)
```

## hue

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species')
```

markers

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species',
           markers=['o', 'x', 'd'])
```

## Small Multiples (FacetGrid)


col

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species',
           col='species')
```

row

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species',
           col='species',
           row='sex')
```

## Extras

x_bins

```python
sns.lmplot(x='culmen_depth_mm', y='culmen_length_mm', 
           data=df[df.species=='Chinstrap'],
           x_bins=5
           )
```
