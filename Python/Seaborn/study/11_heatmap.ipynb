```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
 
```

heatmap map values in a tabular format and each values into particular color so we can find theinsights quicker

```python
cars=sns.load_dataset('mpg').dropna()
```

# basics

```python
#grouping up data
cars.groupby('origin').cylinders.value_counts()
```

```python
orign_cyl=(cars
 .groupby('origin')
 .cylinders
 .value_counts()
 .unstack()
 .fillna(0)
 )
```

```python
sns.heatmap(orign_cyl)
```

```python
#sample one
test_arr=np.array([[1,2],[3,4]])
sns.heatmap(test_arr)
```

Transpose

```python
orign_cyl.T
```

```python
sns.heatmap(orign_cyl.T)
```

Correlation

```python
x=orign_cyl.corr()
x
```

```python
sns.heatmap(x)
```

## collor  palette
the default one is darker color for lower values snd lighter for higher values

```python
#cmap
sns.heatmap(orign_cyl,cmap='Blues')
```

```python
sns.heatmap(orign_cyl,cmap='RdBu')
```

```python
#change the center value
sns.heatmap(orign_cyl,cmap='RdBu',center=60)
```

```python
#mapping maximum and minimum values
sns.heatmap(orign_cyl,cmap='RdBu',vmin=30,vmax=90)
```

## annotations

```python
sns.heatmap(orign_cyl,cmap='RdBu',annot=True)
```

```python
#format the annotations
sns.heatmap(orign_cyl,cmap='RdBu',annot=True,fmt='0f')
```

```python
sns.heatmap(orign_cyl,cmap='RdBu',annot=True,fmt='0.1f')
```

style annotations

```python
#annot_kwa
sns.heatmap(orign_cyl,cmap='RdBu',annot=True,fmt='0f',
            annot_kws={'fontsize':16})
```

```python
sns.heatmap(orign_cyl,cmap='RdBu',annot=True,fmt='0f',
            annot_kws={'fontsize':8,'fontweight':'bold'})
```

```python
sns.heatmap(orign_cyl,cmap='RdBu',annot=True,fmt='0f',
            annot_kws={'fontsize':8,'fontweight':'bold','fontfamily':'serif'})
```

```python
sns.heatmap(orign_cyl,cmap='RdBu',annot=True,fmt='0f',
            annot_kws={'fontsize':8,
                       'fontweight':'bold',
                       'fontfamily':'serif',
                       'color':'black'})
```

## styling

```python
#linewidth
sns.heatmap(orign_cyl,cmap='RdBu',linewidths=5)
```

```python
#linecolor
sns.heatmap(orign_cyl,cmap='RdBu',linewidths=5,
            linecolor='b')
```

Tick labels

```python
xlabels=['A','B','C','D','E']
ylabels=['a','b','c']
sns.heatmap(orign_cyl,cmap='RdBu',
            xticklabels=xlabels,yticklabels=ylabels)
```
