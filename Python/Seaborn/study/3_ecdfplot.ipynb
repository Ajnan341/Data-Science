```python
#add a line at certain dadat eg.4
sns.ecdfplot(x='tip',data=tips)
plt.axvline(4,c='black') #axvline for vertical lines and axhline for horizontal
```

```python
sns.ecdfplot(y='tip',data=tips)
plt.axhline(4,c='black')
```

## additional options

hue: differentiate between categories

```python
sns.ecdfplot(x='tip',data=tips,hue='time')
```

```python
sns.ecdfplot(x='tip',data=tips,hue='day')
```

```python
#palette for styling
sns.ecdfplot(x='tip',data=tips,hue='day',palette='summer')
```

count


```python
sns.ecdfplot(x='tip',data=tips,stat='count')
```

complementary: to know rank of particular value

```python
sns.ecdfplot(x='tip',data=tips,complementary=True,stat='count')# give the opposite of the graph
```

Weights: used to weigh the propotion . Here 50% of the tip money made came from tips less than about $3.50

```python
sns.ecdfplot(x='tip',data=tips,weights='tip')
```
