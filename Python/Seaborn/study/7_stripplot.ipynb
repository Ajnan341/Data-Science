```python
import matplotlib.pyplot as plt
import seaborn as sns
```

steps
1. create a scatter point for each value in the data([4,5,1,2,1,4,7,3])
2. jitter move above or bellow the colliding data points (like 4 and 1 repeated)
it is basically a categorical swarm plot

```python
cars=sns.load_dataset('mpg').dropna()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
cars.head()
```

```python
sns.stripplot(cars.weight)
```

```python
sns.stripplot(x=cars.weight)
```

```python
sns.stripplot(x=cars.origin,y=cars.weight)
```

```python
sns.stripplot(y=cars.origin,x=cars.weight)
```

```python
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders)
```

```python
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True)
```

## jitter

```python
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True,jitter=False)
```

```python
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True,jitter=0.3)
```

```python
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True,jitter=0.7)
```

```python
#alpha
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True,jitter=0.3,
              alpha=0.5)
```

```python
#size
#alpha
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True,jitter=0.3,
              alpha=0.5,size=10)
```

```python
#linewidth and color
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True,
              linewidth=6,edgecolor='r')
```

```python
sns.stripplot(x=cars.origin, y=cars.weight,hue=cars.cylinders,dodge=True,
              linewidth=1,edgecolor='r')
```
