```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
penguins = sns.load_dataset('penguins').dropna()
```

```python
penguins.head()
```

```python
sns.set_style('darkgrid')
```

# basics

```python
sns.barplot(x='species',y='body_mass_g',data=penguins)
```

```python
penguins.groupby('species').body_mass_g.mean()# the height of the barplot is the mean of values of each group 
```

```python
sns.barplot(y='species',x='body_mass_g',data=penguins)
```

It also use bootstrapping to plot the confidence interval

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,errorbar=('ci', 96))
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,
            errorbar=('ci', 16))
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,ci=None)
```

## estimator
helps to change statistics

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,
            estimator=np.std)#standard deviation as estimator
```

```python
#max
sns.barplot(x='species',y='body_mass_g',data=penguins,
            estimator=np.max)#standard deviation as estimator
```

# Categorical variables

hue

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,hue='sex')
```

order: ordering based on requrements

```python
#order
sns.barplot(x='species',y='body_mass_g',data=penguins,
            hue='sex',
            order=['Adelie','Gentoo','Chinstrap'],
            hue_order=['Female','Male'])
```

# Styling

```python
#color
sns.barplot(x='species',y='body_mass_g',data=penguins,color='y')
```

```python
#errcolor and width
sns.barplot(x='species',y='body_mass_g',data=penguins,
            color='y',
            errcolor='c',
            errwidth=10)
```

```python
#errcolor and width
sns.barplot(x='species',y='body_mass_g',data=penguins,
            color='y',
           err_kws={'color': 'c','linewidth': 10})
```

Palatte

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,
            palette='hot')
```

```python
# edge color  & width
sns.barplot(x='species',y='body_mass_g',data=penguins,
            color='aliceblue',
            edgecolor='red',
            lw=6)
```
