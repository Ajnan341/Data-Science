```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('darkgrid')
```

## Basic

```python
sns.pairplot(tips)
```

The pairplot is a combination of different plots default is histograms on diagonals and scatterplot on others


#  Types

```python
#kde as diagonal
sns.pairplot(tips,diag_kind='kde')
```

```python
#hist as diagonal
sns.pairplot(tips,diag_kind='hist')
```

```python
#Change non diagonal plot
sns.pairplot(tips,kind='kde')
```

```python
s=sns.pairplot(tips,kind='reg')
```

```python
#pairgrid
type(s)
```

```python
g=sns.pairplot(tips)
g.map_upper(sns.kdeplot,n_levels=6,color='xkcd:wine red')
```

## Categorical values

```python
tips['weekend']=tips.day.isin(['sat','sun']).fillna(0)
tips.head()
```

hue

```python
sns.pairplot(tips,hue='sex')
```

# vars,xvars and yvars
to specify the attibutes in axis

```python
sns.pairplot(tips,hue='sex',vars=['total_bill','tip'])
```

```python
sns.pairplot(tips,hue='sex',x_vars=['total_bill','tip'],y_vars=['total_bill'])
```

```python

del tips['weekend']
```

# Styling

height,aspect

```python
from ipywidgets import interactive

def size_widget(height=2.5, aspect=1):
    sns.pairplot(tips, height=height, aspect=aspect)
```

```python
interactive(size_widget, height=(1, 3.5, 0.5), aspect=(0.5, 2, 0.25))
```

Palette

```python
sns.pairplot(tips,hue='day',palette='plasma')
```


diag_kws, plot_kws

```python
sns.pairplot(tips, diag_kws={'color': 'gray', 'alpha': 0.2})
```

```python
sns.pairplot(tips, 
             kind='reg',
             plot_kws={'ci':None, 'color': 'xkcd:salmon',
                       'scatter_kws': {'color': 'gray', 's': 5}
                      }
            )
```
