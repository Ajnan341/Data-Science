```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

It is basically categorical plot that you can access all categorical plots like box plot , violinplot ,boxplot, stripplot, barplot, countplot , pointplot, swarmplot.default catplot is a stripplot

```python
#load car data and drop rows with missings
cars = sns.load_dataset('mpg').dropna()

#filter to 4-, 6-, 8-cylinder cars
cars = cars[cars.cylinders.isin([4, 6, 8])] 

#add old/new column
cars['type'] = ['old' if x <=76 else 'new' for x in cars.model_year]
```

```python

cars.shape
sns.set_style('white')
```


## Intro Visuals

```python
sns.catplot(x='horsepower', data=cars)
```

```python
sns.catplot(y='horsepower', data=cars)
```

## Categorical Variables, Hue

```python
sns.catplot(x='origin', y='horsepower', 
            data=cars, 
            kind='box')
```

## hue

```python

sns.catplot(x='origin', y='horsepower', 
            data=cars, 
            hue='cylinders',
            kind='box')#box
```

```python
sns.catplot(x='origin', y='horsepower',
            data=cars, 
            hue='cylinders',
            kind='strip')#strip
```

## Small Multiples (FacetGrid)

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin'
           )#it create diierent plots or each orgin value
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin'
           )
```

## row and column

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            row='type'
           )
```

row_order (order, hue_order, col_order)

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            row='type',
            row_order=['new', 'old'] #type new on top and old on bottom order
           )
```

## Styling
In general, additional keyword arguments will be passed to the kind of plot selected.

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin',
            palette='winter', 
            whis=1 #wisker length (line in the boxplot )
           )
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            s=12,#increaze size of dots
            alpha=0.3,#transparancy
            linewidth=2 #line size of dots
           )
```

## FacetGrid Styling

```python
g = sns.catplot(x='cylinders', y='horsepower', 
                data=cars,
                col='origin',
                kind='box'
               )

type(g)
```

```python
sns.set(style='darkgrid', font_scale=1.5)  

g = sns.catplot(x='cylinders', y='horsepower', 
                data=cars,
                col='origin',
                kind='box'
               )
g.set_titles(col_template="{col_name} cars")
```
