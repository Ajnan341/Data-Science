```python
import matplotlib.pyplot as plt
import seaborn as sns
```

creating by ploting each value in the data it is a relationalplot that give the relation ship between each dataponts

```python
diamd1=sns.load_dataset('diamonds')
```

```python
diamd1.shape
diamd1.head()
```

```python
diamd1.info()
```

```python
diamd1.describe()
```

```python
diamd=diamd1[diamd1.cut.isin(['Premium','Good']) & diamd1.color.isin(['D','F','J'])].sample(n=100,random_state=22)
```

```python
diamd
```

```python
sns.set_style('dark')
```

## BasiCs

```python
sns.scatterplot(x=diamd.carat,y=diamd.price)
```

HUE

```python
sns.scatterplot(x=diamd.carat,y=diamd.price,hue=diamd.cut)
```

```python
sns.scatterplot(x='carat',y='price',data=diamd,hue='cut')
```

```python
# palatte: changing the color of the points by passing a list of colors
sns.scatterplot(x='carat',y='price',data=diamd,hue='cut',palette=['purple','black','red','yellow','blue'])
```

continues variables

```python
sns.scatterplot(x='carat',y='price',data=diamd,hue='depth',palette=['purple','black','red','yellow','blue'])
```

## style

```python
##cut instead of changing color it change the marker
sns.scatterplot(x='carat',y='price',data=diamd,hue='cut'
                ,style='cut'
                )
```

```python
#chage size of marker
sns.scatterplot(x='carat',y='price',data=diamd,hue='cut'
                ,style='cut',s=300
                )
```

size adjust the size of the cut or the points instead of using different color or markers

```python

sns.scatterplot(x='carat',y='price',data=diamd,hue='cut'
                ,size='cut'
                )
```

```python
##also define the size of the cut
sns.scatterplot(x='carat',y='price',data=diamd,hue='cut'
                ,size='cut',sizes=[150,250,300,50,20]
                )
```

```python
#alpha trasperance measure
sns.scatterplot(x='carat',y='price',data=diamd,hue='cut'
                ,size='cut',alpha=0.2
                )
```
