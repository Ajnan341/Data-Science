```python
import matplotlib.pyplot as plt
import seaborn as sns
```

step:
1. sort the values
2. grouping it by values, ([1,1,1,1,2,2,3,4,4,4,5,5,7])
3. place dot for each data point(if above have 4 one so place 4 dots in one's place).
4. do it for all points



```python
cars=sns.load_dataset('mpg').dropna()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
cars.head()
```

```python
sns.swarmplot(cars.horsepower)
```

```python
sns.swarmplot(x=cars.horsepower)
```

```python
#both x and Y 
sns.swarmplot(x=cars.origin,y=cars.horsepower)
```

```python
sns.swarmplot(x='origin', y='horsepower',data=cars)
```

## hue

```python
sns.swarmplot(x='origin', y='horsepower',data=cars,hue='cylinders')
```

```python
#Dodge
sns.swarmplot(x='origin', y='horsepower',data=cars,hue='cylinders',dodge=True)
```

ploting swarm plot over box plot

```python
usa=cars[cars.origin=='usa']
sns.boxplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower)
```

```python
usa=cars[cars.origin=='usa']
sns.boxplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black')
```

```python
#alpha adjust the transparence of dots
sns.boxplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black',alpha=0.4)
```

swarmplot over violinplot

```python
#size of points
sns.violinplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black',alpha=0.4,size=40)
```

```python
#linewidth and edge color
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black',linewidth=2)
```

```python
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black',linewidth=2,edgecolor='red')
```
