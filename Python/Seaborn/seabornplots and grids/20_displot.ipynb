```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

It is distribution plot that contain histplot, kdeplot,ecdfplot and can add rugplot to any of these

```python
cars = sns.load_dataset('mpg').dropna()

cars.head()
```

```python

sns.set_style('white')
```

## Basics

```python
sns.displot(x='weight',data=cars)# here doesnot have the kdeline ontop of histogram
```

```python
# add kde
sns.displot(x='weight', data=cars, kde=True)
```

## kind

## 1. kde(Kernel Density Estimation Plot)

```python
sns.displot(x='weight', data=cars, kind='kde')
```

## 2. ecdf(Empirical Cumulative Distribution Function Plot)

```python
sns.displot(x='weight', data=cars, kind='ecdf')
```

## adding rugplot

```python
sns.displot(x='weight', data=cars, kind='kde', rug=True)
```

## Categorical Variables

1. Hue

```python
sns.displot(x='weight', data=cars, hue='origin')
```

```python
sns.displot(x='weight', data=cars, kind='kde', hue='origin')
```

## Bivariate (2D) Options - NEW!
Displot allow you to plot two variables at once it is call the bivariate or 2D option

```python
sns.displot(x='weight',y='horsepower' ,data=cars)
```

```python
#kde
sns.displot(x='weight', y='horsepower',data=cars, kind='kde')
```

```python
# adding rugplot
sns.displot(x='weight', y='horsepower', data=cars, kind='kde', rug=True)
```

## Small Multiples (FacetGrid) - NEW!
we can now create small multiple plots

Column

```python
sns.displot(x='weight', data=cars, hue='origin'
            , col='origin')
```

```python
sns.displot(x='weight', data=cars, 
            col='origin',
              kind='kde', rug=True)
```

Rows

```python
sns.displot(x='weight', data=cars[cars.cylinders.isin([4, 6, 8])], 
            row='cylinders', 
            hue='cylinders'
           )
```

```python
sns.displot(x='weight', data=cars[cars.cylinders.isin([4, 6, 8])], 
            col='origin', 
            row='cylinders', 
            hue='cylinders'
           )
```

## Styling

palette

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest');
```

height

```python
sns.displot(x='weight', data=cars, col='origin', height=2)
```

```python
sns.displot(x='weight', data=cars, col='origin', height=15)
```

aspect

```python
sns.displot(x='weight', data=cars, col='origin', height=2, aspect=4.5)
```

```python
sns.displot(x='weight', data=cars, col='origin', height=2, aspect=.5)
```

## Other Keywords: Plot Specific
In general, additional keyword arguments will be passed to the kind of plot selected.

```python
sns.displot(x='weight', data=cars, bins=20)#there are 20 bins in teh histogram
```

Limewidth

```python
sns.displot(x='weight', data=cars, kind='kde', fill=True, lw=8)
```

## rug_kws

```python
sns.displot(x='weight', data=cars, kind='kde',
             rug=True, 
             rug_kws={'height': 0.05})
```

## FacetGrid Styling

```python
sns.displot(x='weight', data=cars, col='origin',
             facet_kws={'despine': False})
```

```python
sns.displot(x='weight', data=cars, col='origin',
             facet_kws={'despine': True})
```
