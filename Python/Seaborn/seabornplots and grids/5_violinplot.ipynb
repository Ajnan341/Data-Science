```python
import matplotlib.pyplot as plt
import seaborn as sns
```

It is similar to boxplot it contain median ,25% and 75%.

```python
cars=sns.load_dataset('mpg').dropna()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
cars.head()
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.displacement)
```

```python
sns.violinplot(x='cylinders',y='displacement',data=cars)
```

Hue

```python
sns.violinplot(x='cylinders',y='displacement',data=cars,hue='origin')
```

```python
sns.violinplot(x='cylinders',y='displacement',data=cars[cars.origin.isin(['japan','europe'])],hue='origin')
```

Split: merge the 2 cataegories to singleplot(japan,europe)

```python
sns.violinplot(x='cylinders',y='displacement',data=cars[cars.origin.isin(['japan','europe'])],hue='origin',
               split=True)
```

```python
#inner changing the inner part of the plot

sns.violinplot(x='cylinders',y='displacement',data=cars[cars.origin.isin(['japan','europe'])],hue='origin',
               split=True,
               inner='quartiles')
```

```python
#scale the data using density_norm

sns.violinplot(x='cylinders',y='displacement',data=cars[cars.origin.isin(['japan','europe'])],hue='origin',
               split=True,
               inner='quartiles',
               density_norm='count')
```

```python
#scale the hue
sns.violinplot(x='cylinders',y='displacement',data=cars[cars.origin.isin(['japan','europe'])],hue='origin',
               split=True,
               inner='quartiles',
               density_norm='count',common_norm=False)
```

## styling

```python
#order
sns.violinplot(x='origin',y='displacement',data=cars,order=['japan','europe','usa'])
```

```python
sns.violinplot(x='origin',y='mpg',data=cars,color='g') 
```

```python
sns.violinplot(x='origin',y='mpg',data=cars,hue='origin',width=0.2) 
```

```python
#line width and color
sns.violinplot(x='origin',y='mpg',data=cars,hue='origin',width=0.2,linewidth=2.8,linecolor='g') 
```

```python
sns.violinplot(x='origin',y='mpg',data=cars,hue='origin',width=0.2,bw_adjust=0.2) 
```
