```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

Relplot used to show relationship between variables.Using it we can access the relational plots of seaborn such as scatterplot and lineplot .

```python
df = sns.load_dataset('titanic')

df.head()
```

```python

df['decade'] = [age//10*10 for age in df.age]

df.head()
```

```python
sns.set_style('white')
```

## Basics

```python
sns.relplot(x='age',y='fare',data=df) #default relplot is a scattorplot
```

## Categorical Variables: Hue, Style, Size

Hue

```python
sns.relplot(x='age',y='fare',data=df,hue='pclass')
```

## Style
put different markers

```python
sns.relplot(x='age',y='fare',data=df,
            style='pclass')
```

```python
sns.relplot(x='age',y='fare',data=df,
            hue='pclass',
            style='pclass')
```

## Size
diff size of markers

```python
sns.relplot(x='age',y='fare',data=df,
            size='pclass')
```

```python
sns.relplot(x='age',y='fare',data=df,
            hue='pclass',
            style='pclass',
            size='pclass')
```

## kind

1. Lineplot

```python
sns.relplot(x='age',y='fare',data=df,
            kind='line')
```

```python
sns.relplot(x='age',y='fare',data=df,
            hue='pclass',
            kind='line')
```

```python
sns.relplot(x='decade', y='fare', data=df, kind='line')
```

```python
sns.relplot(x='decade', y='fare', 
            data=df, 
            kind='line',
            hue='pclass',
            style='pclass'
           )
```

## Small Multiples (FacetGrid)

column

```python
sns.relplot(x='age', y='fare',
            data=df,
            col='pclass'
           )
```

```python
sns.relplot(x='decade', y='fare',
            data=df,
            kind='line',
            col='pclass'
           )
```

Row

```python
sns.relplot(x='decade', y='fare',
            data=df,
            kind='line',
            row='pclass'
           )
```

```python
sns.relplot(x='age', y='fare',
            data=df,
            col='deck'
           )
```

the above plot deck have 7 classes its is very small and arrangement is very poor .Therefore wecan use col_wrap that help to arrange better ie below it in a way that each row contai 3 plots.

```python
sns.relplot(x='age', y='fare',
            data=df,
            col='deck',
            col_wrap=3
           )
```

```python
sns.relplot(x='age', y='fare',
            data=df,
            col='deck',
            col_wrap=2
           )
```

## Styling
In general, additional keyword arguments will be passed to the kind of plot selected.

```python
#increaze size of dots
sns.relplot(x='age', y='fare', data=df,
            col='pclass',
            hue='survived',
            s=150)
```

palette

```python
sns.relplot(x='age', y='fare', data=df,
            col='pclass',
            hue='survived',
            palette=['black', 'red']
           )
```

change marker

```python

sns.relplot(x='age', y='fare', data=df,
            col='pclass',
            hue='survived',
            marker='^'
           )
```

```python

sns.relplot(x='age', y='fare', data=df,
            col='pclass',
            hue='survived',
            s=150,
            palette=['black', 'red'],
            marker='^'
           )
```

```python
sns.relplot(x='age', y='fare', data=df,
            col='pclass',
            hue='survived',
            s = 150, 
            style='survived'
            
           )
```

```python
sns.relplot(x='age', y='fare', data=df,
            col='pclass',
            hue='survived',
            s = 150, 
            palette=['black', 'red'],
            style='survived',
            markers=['X', 'o']#two markers
           )
```

## FacetGrid Styling

```python
g = sns.relplot(x='decade', y='survived',
                data=df[df.age < 70],
                kind='line',
                hue='sex',
                ci=None
               )
```

```python
g = sns.relplot(x='decade', y='survived',
                data=df[df.age < 70],
                kind='line',
                hue='sex',
            errorbar=None,
                col='pclass'
               )

g.set_xlabels('Passenger Age Group')
g.set_ylabels("Percent Survived")
```
