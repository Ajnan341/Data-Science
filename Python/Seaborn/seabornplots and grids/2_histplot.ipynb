```python
import seaborn as sns
sns.__version__
```

```python
peng=sns.load_dataset('penguins')
```

```python
peng.info()
```

```python
peng.describe()
```

```python
peng.dropna(inplace=True)
```

```python
peng
```

## basic

```python
sns.histplot(peng.flipper_length_mm)
```

```python
#anotherway
sns.histplot(x='flipper_length_mm',data=peng)
```

```python
#horizontal
sns.histplot(y='flipper_length_mm',data=peng)
```

```python
#distplot (older) ie Histplot+kde
sns.histplot(x='flipper_length_mm',data=peng,kde=True)
```

BIN values

```python
sns.histplot(x='flipper_length_mm',data=peng,bins=20)#20 bins
```

```python
# arrage locations of bins to appear(starting and ending locations of bins)

sns.histplot(x='flipper_length_mm',data=peng,bins=[180,190,220,230])
```

binwidth:define the size or width of bins

```python
sns.histplot(x='flipper_length_mm',data=peng,binwidth=10)#each bin have a width of 0 units
```

binrange defining the range of bins

```python
sns.histplot(x='flipper_length_mm',data=peng,binwidth=10,binrange=(180,200))
```

```python
sns.histplot(x='flipper_length_mm',data=peng,binrange=(180,200))
```

## Statics

```python
#Density
sns.histplot(x='flipper_length_mm',data=peng,stat='density',color='r')
```

```python
#probability : height of histogram bars to one
sns.histplot(x='flipper_length_mm',data=peng,stat='probability')
```

## categorical variables

hue: pass the name of coulmn that contain categorical data

```python
sns.histplot(x='flipper_length_mm',data=peng,hue='species')
```

element: to avoid overlaping

```python
#steps
sns.histplot(x='flipper_length_mm',data=peng,hue='species',element='step')
```

```python
#poly
sns.histplot(x='flipper_length_mm',data=peng,hue='species',element='poly')
```

multiple

```python
#layer :default
sns.histplot(x='flipper_length_mm',data=peng,hue='species',multiple='layer')
```

```python
#stack
sns.histplot(x='flipper_length_mm',data=peng,hue='species',multiple='stack')
```

```python
#fill
sns.histplot(x='flipper_length_mm',data=peng,hue='species',multiple='fill')
```

## bivariate hist plot

```python
peng
```

```python
sns.histplot(x='bill_length_mm',y='bill_depth_mm',data=peng)
```

```python
#colorbar
sns.histplot(x='bill_length_mm',y='bill_depth_mm',data=peng,cbar=True)
```

```python
sns.histplot(x='bill_length_mm',y='bill_depth_mm',data=peng,cbar=True,hue='species')
```

## Styling

```python
#dodge& shrink
sns.histplot(x='species',hue='sex',multiple='dodge',data=peng,shrink=0.2)
```

```python
#color
sns.histplot(x='species',data=peng,color='g')
```

```python
sns.histplot(x='species',hue='sex',multiple='dodge',data=peng,shrink=0.8, palette='bone')
```

```python
sns.histplot(x='species',data=peng,fill=False)
```
