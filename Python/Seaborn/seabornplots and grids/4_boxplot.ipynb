```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Steps
1. find the median 
2. find 25% of data at left of median 
3. find 25% of data at right(75%)
4. draw lines 
5. then connect the line s and form the box portion
6. next step is to build the whiskers for that find the  inner Quartile range(IQR) ie it is the distance from 25th percentile to 75 percentile.
7. after finding the IQR stretch the whiskers out in either 1.5*IQR in both directions.
8. end the whiskers a the last point within the 1.5*IQR range(less than or equal) 

using it you can split upthe data into different category(more then one box plot) and identify outliers in data

```python
cars=sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
cars.head()
```

```python
sns.boxplot(cars.mpg)
```

```python
sns.boxplot(x=cars.mpg)
```

```python
cars.mpg.describe()
```

more than one attributes

```python
sns.boxplot(x=cars.origin,y=cars.mpg) # the circles are outliers
```

```python
sns.boxplot(x='origin',y='mpg',data=cars) # the circles are outliers
```

## hue

```python
sns.boxplot(x='origin',y='mpg',data=cars,hue='cylinders') # the circles are outliers
```

```python
cars['newer_model']=cars.model_year>76
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model') 
```

## Sytling options

```python
#set order
sns.boxplot(x='origin',y='mpg',data=cars,order=['japan','europe','usa'])
```

```python
#hue order
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',hue_order=[True,False]) 
```

```python
#color
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',color='g') 
```

```python
# width of box bar
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',width=0.2) 
```

```python
#linewidth and color
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',linewidth=2.8,linecolor='g') 
```

```python
#whiskers range
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',whis=.5) 
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',whis=3) 
```

```python
#flier size(outlier)
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',fliersize=2) 
```

```python
#showcaps
sns.boxplot(x='origin',y='mpg',data=cars,hue='newer_model',showcaps=False) 
```
