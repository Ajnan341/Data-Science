```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

Lmplot stands for linear model plot with it we can create linear models for data like regplot and also has facetgrit to create multiple regplots

```python
df = sns.load_dataset('penguins').dropna()

df.head()
```

```python
sns.set_style('white')
```

# Basic


```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df)
```

## hue

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species')
```

markers

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species',
           markers=['o', 'x', 'd'])
```

## Small Multiples (FacetGrid)


col

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species',
           col='species')
```

row

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df
           ,hue='species',
           col='species',
           row='sex')
```

## Extras

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', 
           data=df[df.species=='Chinstrap']
           )
```

x_bins

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', 
           data=df[df.species=='Chinstrap'],
           x_bins=5
           )# the line contain the avg values of that bucket
```

## Other models: Logistic Regression,
logistic

```python
df['sex_binary'] = [(x == 'Male') for x in df.sex]
```

```python

df.head()
```

```python
sns.lmplot(x='body_mass_g', y='sex_binary', data=df, 
           logistic=True, # logistic regression
           n_boot=100 #specify the no bootstraps
           )
```

## Styling
scatter_kws

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', 
           data=df, 
           hue='species',
           col='species',
           scatter_kws={'s': 150, 'edgecolor': 'black'}
           )
```

line_kws

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', 
           data=df, 
           hue='species',
           col='species',
           line_kws={'lw': 5, 'ls': '-.', 'color': 'black'}
           )
```

## FacetGrid Object

```python

g = sns.lmplot(x='bill_length_mm', y='bill_depth_mm', 
               data=df, 
               hue='species', 
               col='species'
              )
```

```python
g = sns.lmplot(x='bill_length_mm', y='bill_depth_mm', 
               data=df, 
               hue='species', 
               col='species'
              )

g.set_titles(col_template='{col_name} Penguins')
g.set_xlabels('Culmen Depth (mm)')
```
