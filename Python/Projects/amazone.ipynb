```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

service = Service(executable_path=r"D:\1mini project\chromedriver-win64 (2)\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service)

def get_amazon_price(product):
    driver.get('https://www.amazon.in')

    # Wait until the search box is present
    search_box = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "twotabsearchtextbox"))
    )
    
    # Enter the product name and hit enter
    search_box.send_keys(product)
    search_box.send_keys(Keys.RETURN)
    
    # Wait until the search results are present
    search_results = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, ".s-main-slot .s-result-item"))
    ) 
    
    print('search results')
    print(search_results)
    
    product_names = []
    product_prices = []

    # Extract product names and prices
    products = driver.find_elements(By.CSS_SELECTOR, ".s-main-slot .s-result-item")
    print(products)
    count = 0
    for product in products:
        print(product)
        if count >= 10:
            break
        try:
            name = product.find_element(By.CSS_SELECTOR, "h2 a span").text
            price = product.find_element(By.CSS_SELECTOR, ".a-price-whole").text
            product_names.append(name)
            product_prices.append(price)
            count += 1
        except:
            continue
    
    # Close the driver
    driver.quit()
    
    # Return product names and prices
    return list(zip(product_names, product_prices))

# Example usage
product_details = get_amazon_price("iphone  15")
for name, price in product_details:
    print(f"Product: {name}, Price: ₹{price}")
```
