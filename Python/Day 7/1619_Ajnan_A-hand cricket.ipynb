```python
import random
import math

class Toss():
    def __init__(self,ip):
        self.ip=ip.upper()
        self.cpu_choice = 0
        self.player_choice = 0

    def conduct_toss(self):
        self.cpu_choice = random.randint(1, 6)
        self.player_choice = int(input("Toss your number (1-6): "))
        print(f"TOSS : YOU : {self.player_choice}")
        print(f"TOSS : COMPUTER : {self.cpu_choice}")


    def determine_status(self):

        if self.ip == "EVEN":
            if (self.cpu_choice + self.player_choice) % 2 == 0:
                return input("Bowling or Fielding? (BATTING/BOWLING): ").upper()
            else:
                return self.computer_choice()
            
        elif self.ip == "ODD":
            if (self.cpu_choice + self.player_choice) % 2 == 1:
                return input("Bowling or Fielding? (BATTING/BOWLING): ").upper()
            else:
                return self.computer_choice()
            
    def computer_choice(self):
        choice = random.random()
        if choice < 0.5:
            print("Computer won the toss and chose to BAT.")
            return "BOWLING"
        else:
            print("Computer won the toss and chose to BOWL.")
            return "BATTING"
        
        
class Innings:
    def __init__(self, status):
        self.status = status
        self.score = 0
        self.target = 1
        self.innings = 1

    def play(self):
        while True:
            player_input = int(input(f"You: {self.status} innings {self.innings} (Enter your number): "))
            cpu_input = random.randint(1, 6)
            print(f"You: {player_input}")
            print(f"Computer: {cpu_input}")

            if player_input == cpu_input:
                if self.status == "BATTING" and self.innings == 1:
                    self.status = "BOWLING"
                    self.innings = 2
                    print("OUT... TO DEFEND: " + str(self.score))
                elif self.status == "BOWLING" and self.innings == 1:
                    self.status = "BATTING"
                    self.innings = 2
                    print("OUT... TO GET: " + str(self.target))
                elif self.status == "BATTING" and self.innings == 2 and self.target>self.score:
                    print("OUT... COMPUTER WON THE MATCH")
                    break
                elif self.status == "BOWLING" and self.innings == 2 and self.target<self.score:
                    print("OUT... YOU WON THE MATCH!")
                    break
            else:
                self.update_scores(player_input, cpu_input)

    def update_scores(self, player_input, cpu_input):
        if self.status == "BATTING":
            if self.innings == 1:
                self.score += player_input
                print(f"SCORE: {self.score}")
            elif self.innings == 2:
                if self.target - player_input > 0:
                    self.target -= player_input
                    print(f"TO GET: {self.target}")
                else:
                    print("YOU WON THE MATCH!")
                    return
        elif self.status == "BOWLING":
            if self.innings == 1:
                self.target += cpu_input
                print(f"SCORE: {self.target}")
            elif self.innings == 2:
                if self.score - cpu_input > 0:
                    self.score -= cpu_input
                    print(f"TO DEFEND: {self.score}")
                else:
                    print("YOU LOST THE MATCH :(")
                    return
                

class Game:
    def __init__(self):
        selected = input("ODD or EVEN? ").upper()
        self.toss = Toss(selected)

    def start(self):
        self.toss.conduct_toss()
        status = self.toss.determine_status()
        innings = Innings(status)
        innings.play()


if __name__ == "__main__":
    game = Game()
    game.start()

    
```
