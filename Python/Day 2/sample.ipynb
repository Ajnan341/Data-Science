```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

service = Service(executable_path=r"D:\1mini project\chromedriver-win64 (2)\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service)

def get_flipkart_price(product):
    try:
        driver.get('https://www.flipkart.com')

        # Handle login popup if present (optional)
        try:
            close_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "button._2KpZ6l._2doB4z"))
            )
            close_button.click()
        except TimeoutException:
            pass
        
        # Wait until the search box is present and clickable
        try:
            search_box = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "input[type='text'][title='Search for products, brands and more']"))
            )
        except TimeoutException:
            print("Timeout occurred while waiting for search box.")
            driver.quit()
            return []

        # Enter the product name and hit enter
        search_box.clear()
        search_box.send_keys(product)
        search_box.submit()
        
        # Wait until the search results are present
        try:
            WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div._1AtVbE"))
            )
        except TimeoutException:
            print("Timeout occurred while waiting for search results.")
            driver.quit()
            return []

        product_names = []
        product_prices = []

        # Extract product names and prices
        products = driver.find_elements(By.CSS_SELECTOR, "div._1AtVbE")
        
        count = 0
        for product in products:
            if count >= 10:
                break
            try:
                name = product.find_element(By.CSS_SELECTOR, "a.s1Q9rs").text
                price = product.find_element(By.CSS_SELECTOR, "div._30jeq3").text
                product_names.append(name)
                product_prices.append(price)
                count += 1
            except:
                continue
        
        # Close the driver
        driver.quit()
        
        # Return product names and prices
        return list(zip(product_names, product_prices))
    
    except Exception as e:
        print(f"An error occurred: {e}")
        driver.quit()
        return []

# Example usage
product_details = get_flipkart_price("laptop")
for name, price in product_details:
    print(f"Product: {name}, Price: ₹{price}")

```
