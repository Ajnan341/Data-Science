```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

service = Service(executable_path=r"D:\1mini project\chromedriver-win64 (2)\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service)
def get_amazon_price(product):
    driver.get('https://www.flipkart.com/')
    
    try:
        # Use WebDriverWait to wait for the element to be present
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, 'twotabsearchtextbox'))
        )
        search_box.send_keys(product)
        search_box.submit()
        
        # Wait for the price element to be present
        price = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//span[@class="a-price-whole"]'))
        ).text
        cents = driver.find_element(By.XPATH, '//span[@class="a-price-fraction"]').text
        price = f"{price}.{cents}"
    except:
        price = 'Not found'
    
    return price

def get_ebay_price(product):
    driver.get('https://www.ebay.com')
    
    try:
        # Use WebDriverWait to wait for the element to be present
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//input[@type="text" and @name="_nkw"]'))
        )
        search_box.send_keys(product)
        search_box.submit()
        
        # Wait for the price element to be present
        price = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//span[@class="s-item__price"]'))
        ).text
    except:
        price = 'Not found'
    
    return price

def compare_prices(product):
    amazon_price = get_amazon_price(product)
    ebay_price = get_ebay_price(product)
    
    print(f"Prices for '{product}':")
    print(f"Amazon: ${amazon_price}")
    print(f"eBay: {ebay_price}")
    
    driver.quit()

if __name__ == "__main__":
    product = "IPHONE 15"  # Replace with the product you want to search for
    compare_prices(product)
```
