```python
from tkinter import *


t = Tk()
t.title('Calculator')
t.geometry('267x270')
t.config(bg='navy')


entry = Entry(t, width=39, borderwidth=15, bg='coral')
entry.grid(row=0, column=0, columnspan=5)

# Function to update the Entry widget
def click_button(value):
    #print(value)
    current_text = entry.get()
    #print('before :',entry.get())
    entry.delete(0, END)
    #print('after ',entry.get())
    #print(current_text + value)
    entry.insert(0, current_text + value)

# Function to clear the Entry widget
def clear_entry():
    entry.delete(0, END)




# perform evaluations
def evaluate_expression():
    try:
        result = eval(entry.get())
        #print(entry.get())
        #print(result)
        entry.delete(0, END)
        entry.insert(0, str(result))
    except Exception as e:
        entry.delete(0, END)
        entry.insert(0, "Error")


#button format
bt1 = Button(t, text='1', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('1'))
bt1.grid(row=2, column=0)

bt2 = Button(t, text='2', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('2'))
bt2.grid(row=2, column=1)

bt3 = Button(t, text='3', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('3'))
bt3.grid(row=2, column=2)

bt4 = Button(t, text='4', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('4'))
bt4.grid(row=3, column=0)

bt5 = Button(t, text='5', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('5'))
bt5.grid(row=3, column=1)

bt6 = Button(t, text='6', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('6'))
bt6.grid(row=3, column=2)

bt7 = Button(t, text='7', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('7'))
bt7.grid(row=4, column=0)

bt8 = Button(t, text='8', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('8'))
bt8.grid(row=4, column=1)

bt9 = Button(t, text='9', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('9'))
bt9.grid(row=4, column=2)

bt0 = Button(t, text='0', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('0'))
bt0.grid(row=5, column=0)

btplus = Button(t, text='+', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('+'))
btplus.grid(row=2, column=3)

btminus = Button(t, text='-', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('-'))
btminus.grid(row=3, column=3)

btmultiply = Button(t, text='*', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('*'))
btmultiply.grid(row=4, column=3)

btdivide = Button(t, text='/', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('/'))
btdivide.grid(row=5, column=3)

btequal = Button(t, text='=', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=evaluate_expression)
btequal.grid(row=5, column=2)

btclear = Button(t, text='Clear', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=clear_entry)
btclear.grid(row=5, column=1)

btDecimal = Button(t, text='.', font=("Arial Bold", 10), bg='black', fg='white', height=2, width=7, command=lambda: click_button('.'))
btDecimal.grid(row=6, column=0)


t.mainloop()
```
