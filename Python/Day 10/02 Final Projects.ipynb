```python
import tkinter as tk
import random

def load_questions(filename):
    questions = []
    answers = []
    with open('question ans.csv', 'r') as f:
        for line in f:
            question, answer = line.strip().split(',')
            questions.append(question)
            answers.append(answer)
    return questions, answers

def generate_quiz(questions, num_questions):
    quiz = random.sample(questions, num_questions)
    return quiz

def create_gui(quiz, answers):
    root = tk.Tk()
    root.title("Quiz")

    # Create labels and entry fields for questions and answers
    question_labels = []
    answer_entries = []
    for i, question in enumerate(quiz):
        label = tk.Label(root, text=f"Question {i+1}: {question}")
        label.pack()
        entry = tk.Entry(root)
        entry.pack()
        question_labels.append(label)
        answer_entries.append(entry)

    def submit():
        student_answers = [entry.get() for entry in answer_entries]
        score = grade_quiz(student_answers, answers)
        result_label = tk.Label(root, text=f"Your score: {score}/{len(quiz)}")
        result_label.pack()

    submit_button = tk.Button(root, text="Submit", command=submit)
    submit_button.pack()

    root.mainloop()

def grade_quiz(student_answers, answers):
    correct = 0
    for i in range(len(student_answers)):
        if student_answers[i] == answers[i]:
            correct += 1
    return correct

if __name__ == "__main__":
    questions_file = "questions.txt"
    num_questions = 10

    questions, answers = load_questions(questions_file)
    quiz = generate_quiz(questions, num_questions)
    create_gui(quiz, answers)
```

```python
import tkinter as tk
from tkinter import messagebox
import random
import pandas as pd

# Load the CSV file
csv_file_path = 'question ans.csv'
data = pd.read_csv(csv_file_path)

# Convert the DataFrame to a dictionary
q_A_data =  data.to_dict(orient='list')



class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("MCQ Quiz")
        self.score = 0
        self.question_index = 0
        self.total_questions = 6
        self.selected_questions = random.sample(range(len(q_A_data['Questions'])), self.total_questions)

        self.question_label = tk.Label(root, text="", wraplength=500, justify="left", font=("Arial", 14))
        self.question_label.pack(pady=20)

        self.selected_answer = tk.StringVar(value="")

        self.option_a = tk.Radiobutton(root, text="", variable=self.selected_answer, value="A", font=("Arial", 12))
        self.option_a.pack(anchor="w")
        self.option_b = tk.Radiobutton(root, text="", variable=self.selected_answer, value="B", font=("Arial", 12))
        self.option_b.pack(anchor="w")
        self.option_c = tk.Radiobutton(root, text="", variable=self.selected_answer, value="C", font=("Arial", 12))
        self.option_c.pack(anchor="w")
        self.option_d = tk.Radiobutton(root, text="", variable=self.selected_answer, value="D", font=("Arial", 12))
        self.option_d.pack(anchor="w")

        self.next_button = tk.Button(root, text="Next", command=self.next_question, font=("Arial", 12))
        self.next_button.pack(pady=20)

        self.show_question()

    def show_question(self):
        q_index = self.selected_questions[self.question_index]
        self.question_label.config(text=f"Question {self.question_index + 1}: {q_A_data['Questions'][q_index]}")

        self.option_a.config(text=f"A. {q_A_data['Option A'][q_index]}")
        self.option_b.config(text=f"B. {q_A_data['Option B'][q_index]}")
        self.option_c.config(text=f"C. {q_A_data['Option C'][q_index]}")
        self.option_d.config(text=f"D. {q_A_data['Option D'][q_index]}")

        self.selected_answer.set("")

    def next_question(self):
        q_index = self.selected_questions[self.question_index]
        user_answer = self.selected_answer.get()
        correct_answer = self.get_correct_answer(q_index)

        if user_answer and user_answer == correct_answer:
            self.score += 1

        self.question_index += 1

        if self.question_index < self.total_questions:
            self.show_question()
        else:
            self.show_score()

    def get_correct_answer(self, q_index):
        correct_text = q_A_data['Correct answer'][q_index]
        if correct_text == q_A_data['Option A'][q_index]:
            return "A"
        elif correct_text == q_A_data['Option B'][q_index]:
            return "B"
        elif correct_text == q_A_data['Option C'][q_index]:
            return "C"
        elif correct_text == q_A_data['Option D'][q_index]:
            return "D"

    def show_score(self):
        messagebox.showinfo("Quiz Finished", f"Your score is: {self.score}/{self.total_questions}")
        self.root.quit()

# Create the main window
root = tk.Tk()
quiz_app = QuizApp(root)
root.mainloop()
```
