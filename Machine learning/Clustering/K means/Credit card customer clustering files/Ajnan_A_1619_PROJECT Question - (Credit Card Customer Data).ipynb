## <span style = "color:orange"> Visualization </span>

### Plot a scatterplot of Avg_credit_Limit vs Total_Credit_Cards with respect to Total_calls_made

```python
sns.scatterplot(x='Avg_Credit_Limit',y='Total_Credit_Cards',data=df,hue='Total_calls_made')
```

### Plot correlation matix on Heatmap

```python
sns.heatmap(df.corr(),annot=True)
```

### Plot a pairplot

```python
sns.pairplot(df)
```

### Create K Means model (Let k = 4)

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=4)
```

### Fit predict for our Dataset

```python
c=model.fit_predict(df)
c
```

### Assign the predicted clusters to a new column in our dataset

```python
df['clusters']=c
df.head()
```

```python
df.head()
```

```python
df['clusters'].unique()
```

## Visualize the clusters

```python
# Use x = 'Avg_Credit_Limit' ang y =  'Total_Credit_Cards'
df0 = df[df['clusters'] == 0]
df1 = df[df['clusters'] == 1]
df2 = df[df['clusters'] == 2]
df3 = df[df['clusters'] == 3]
```

```python
sns.scatterplot(data = df0, x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', color = 'green')
sns.scatterplot(data = df1, x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', color = 'red')
sns.scatterplot(data = df2, x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', color = 'blue')
sns.scatterplot(data = df3, x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', color = 'yellow')
```
